```python
filename = "./input_1.txt"

with open(file=filename) as f:
    input_data_pt2 = f.read().splitlines()
```

# Pt. 2

```python
position = 50
count = 0
for i in input_data_pt2:
    previous = position
    parity = 1
    if i[0] == "L":
        parity = -1

    carry = 0
    if position == 0 and parity == -1:
        carry = 1

    step = int(i[1:])
    if step == 0:
        print("WARNING!!")
    position = position + parity * step

    top_down = 0
    if position%100 == 0 and position < 1:
        top_down = 1
    
    delta = abs(position//100) - carry + top_down
    count += delta

    position = position%100

    # print(f"{previous:>2}, {position:>2}, {i:>5}, {delta:>3}, {count}")
```
