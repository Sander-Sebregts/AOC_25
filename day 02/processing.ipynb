```python
filename = "./input_1.txt"

with open(file=filename) as f:
    input_data_pt1 = f.read().splitlines()
```

# Pt. 1

```python
ranges = input_data_pt1[0].split(",")
invalids = []
for range_n in ranges:
    [range_start, range_end] = range_n.split("-")

    if len(range_start) % 2 == 0:
        min_value = int(range_start[:len(range_start)//2])
    else:
        min_value = 10**int(len(range_start)/2)

    if len(range_end) % 2 == 0:
        max_value = int(range_end[:len(range_end)//2])
    else:
        max_value = int("9"*int(len(range_end)/2))

    for i in range(min_value, max_value+1):
        value = int(str(i)*2)
        if value >= int(range_start) and value <= int(range_end):
            invalids.append(value)
    # print(range_start, range_end, min_value, max_value)
```

```python
print(sum(invalids), invalids)
```

```python
filename = "./input_1.txt"

with open(file=filename) as f:
    input_data_pt2 = f.read().splitlines()
```

# Pt. 2

```python
ranges = input_data_pt2[0].split(",")
invalids = []
for range_n in ranges:
    [range_start, range_end] = range_n.split("-")

    # Only extension for pt 2: loop over amount of repeating number of digits, use set to drop duplicate values (e.g. 2121 2121 and 21 21 21 21, only need to count 21212121 once)
    range_invalids = set()
    for l in range(2, len(range_end)+1):
    # End of addition for pt 2 apart from replacing "2" with "l"
        # Determine minimum value of repeating number based on number of digits
        if len(range_start) % l == 0:
            min_value = int(range_start[:len(range_start)//l])
        else:
            min_value = 10**int(len(range_start)/l)

        # Determine maximum value of repeating number based on number of digits
        if len(range_end) % l == 0:
            max_value = int(range_end[:len(range_end)//l])
        else:
            max_value = int("9"*int(len(range_end)/l))

        # Loop from min to max value of repeating number
        for i in range(min_value, max_value+1):
            # Repeat number l times
            value = int(str(i)*l)
            # Validate if number between min and max bounds of the range
            if value >= int(range_start) and value <= int(range_end):
                range_invalids.add(value)
    # Extend the list of invalid numbers by the list-made set of this range
    invalids.extend(list(range_invalids))
```

```python
print(sum(invalids), invalids)
```
