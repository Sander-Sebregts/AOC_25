# Pt. 1

```python
def jolts_per_bank(bank: str, n_batteries_per_line: int = 2) -> int:
    first = 0
    first_index = 0
    for index, value in enumerate(bank[:-n_batteries_per_line+1]):
        if int(value) > first:
            first = int(value)
            first_index = index
    second = 0
    for value in bank[first_index+1:]:
        if int(value) > second:
            second = int(value)
    return first*10+second
```

Author's note: yay for being foresightful enough to include n_batteries_per_line (should've called it "per_bank" though...), too bad the implementation was lacking here.

# Pt. 2

```python
def jolts_per_bank(bank: str, n_batteries_per_line: int = 2) -> int:
    found_values = ""
    lowest_index = 0
    for j in range(n_batteries_per_line):
        battery = ""
        sub_index = 0
        if j!= n_batteries_per_line-1:
            enumerator = enumerate(bank[lowest_index:-n_batteries_per_line+j+1])
        else:
            enumerator = enumerate(bank[lowest_index:])
        for index, value in enumerator:
            if value > battery:
                battery = value
                sub_index = index
        found_values += battery
        lowest_index += sub_index + 1
    return int(found_values)
```

```python
jolts = []
for bank in input_data_pt2:
    jolts.append(jolts_per_bank(bank=bank, n_batteries_per_line=12))
sum(jolts)
```
