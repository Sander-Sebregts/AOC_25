# Pt. 1

```python
def extract_ranges_and_items(input: list[str]) -> list[list[list[int]], list[int]]:
    ranges = []
    items = []
    for line in input:
        if "-" in line:
            ranges.append([int(line.split("-")[0]), int(line.split("-")[1])])
        elif line:
            items.append(int(line))
        else:
            continue
    return [ranges, items]

def validate_freshness(ranges: list[list[int]], items: list[int]) -> list[int]:
    fresh = []
    for item in items:
        for [min, max] in ranges:
            if min<=item<=max:
                fresh.append(item)
                break
    return fresh
```

```python
fresh_items = validate_freshness(*extract_ranges_and_items(input=input_data_pt1))
print(f"Found {len(fresh_items)} items to still be fresh!")
```

```python
def combine_ranges(ranges: list[list[int]]) -> list[list[list[int]], bool]:
    sorted_ranges = sorted(ranges, key=lambda x: x[0])
    combined_ranges = sorted_ranges[:1]
    for [r_start, r_end] in sorted_ranges[1:]:
        for index, [start, end] in enumerate(combined_ranges):
            if end>=r_start>=start or end>=r_end>=start:
                combined_ranges[index] = [min(start,r_start), max(end,r_end)]
                break
            if index == len(combined_ranges)-1:
                combined_ranges.append([r_start, r_end])
                break
    return combined_ranges

def get_range_coverage(ranges: list[list[int]]) -> int:
    coverage = 0
    for [start, end] in ranges:
        coverage += end-start+1
    return coverage
```

```python
ranges, _ = extract_ranges_and_items(input=input_data_pt2)
combined_ranges = combine_ranges(ranges=ranges)
print(f"Total coverage of fresh products: {get_range_coverage(combined_ranges)}")
```
