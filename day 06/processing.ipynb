# Pt. 1

```python
def transpose_list_of_lists(listlist: list[list]) -> list[list]:
    new_listlist = []
    for i in range(len(listlist[0])):
        new_listlist.append([x[i] for x in listlist])
    return new_listlist

def calculate_order(order: list) -> int:
    operand = order[-1]
    values = [int(x) for x in order[:-1]]
    if operand == "*":
        return prod(values)
    elif operand == "+":
        return sum(values)
    else:
        print(f"Unknown operand {operand}")
    
```

```python
instructions = []
for line in input_data_pt1:
    instructions.append([x for x in line.split(" ") if x])
logic_instructions = transpose_list_of_lists(instructions)

all_orders = []
for instruction in logic_instructions:
    all_orders.append(calculate_order(order=instruction))

print(f"Sum of all orders is {sum(all_orders)}")
# print(f"Based on orders {logic_instructions}")
```

```python
def read_cephalopod_numbers(listlist: list[list]) -> list[str]:
    bounds = [i for i, value in enumerate(listlist[-1]) if value != " "] + [len(listlist[-1])+1]
    instructions = []
    for number in range(len(bounds[:-1])):
        cephalopod_numbers = []
        for i in range(bounds[number], bounds[number+1]-1):
            cephalopod_numbers.append("".join([l[i] for l in listlist[:-1] if l[i] != " "]))
        instructions.append(cephalopod_numbers+[listlist[-1][bounds[number]]])
    return instructions
```

```python
cephalopod_instructions = read_cephalopod_numbers(input_data_pt2)
all_orders = []
for instruction in cephalopod_instructions:
    all_orders.append(calculate_order(order=instruction))

print(f"Sum of all orders is {sum(all_orders)}")
# print(f"Based on orders {cephalopod_instructions}")
```
