# Pt. 1

```python
def propagate_beam(grid: list[list[str]]) -> list[int, list[list[str]]]:
    n_splits = 0
    for y in range(len(grid)):
        for x in range(len(grid[y])):
            if grid[y][x] == "S":
                grid[y] = grid[y][:x]+"|"+grid[y][x+1:]
                break
            elif grid[y][x] == "^":
                if grid[y-1][x] == "|":
                    n_splits +=1
                    try:
                        if grid[y][x-1] != "^":
                            grid[y] = grid[y][:x-1]+"|"+grid[y][x:]
                    except IndexError:
                        continue
                    try:
                        if grid[y][x+1] != "^":
                            grid[y] = grid[y][:x+1]+"|"+grid[y][x+2:]
                    except IndexError:
                        continue
            else:
                try:
                    if grid[y-1][x] == "|":
                        grid[y] = grid[y][:x]+"|"+grid[y][x+1:]
                except IndexError:
                    continue
    return [n_splits, grid]
```

```python
n_splits, _ = propagate_beam(grid=input_data_pt1)
print(f"Number of times beams are split: {n_splits}")
```

# Pt. 2

```python
def propagate_beams(grid: list[list[str]]) -> list[int, list[list[str]], int, list[list[int]]]:
    beams = [[0 for _ in range(len(grid[0]))] for _ in range(len(grid))]
    n_splits = 0
    for y in range(len(grid)):
        for x in range(len(grid[y])):
            if grid[y][x] == "S":
                grid[y] = grid[y][:x]+"|"+grid[y][x+1:]
                beams[y][x] = 1
                break
            elif grid[y][x] == "^":
                if grid[y-1][x] == "|":
                    n_splits +=1
                    try:
                        if grid[y][x-1] != "^":
                            grid[y] = grid[y][:x-1]+"|"+grid[y][x:]
                            beams[y][x-1] += beams[y-1][x]
                    except IndexError:
                        print("left alignment failure")
                        continue
                    try:
                        if grid[y][x+1] != "^":
                            grid[y] = grid[y][:x+1]+"|"+grid[y][x+2:]
                            beams[y][x+1] += beams[y-1][x]
                    except IndexError:
                        print("right alignment failure")
                        continue
            else:
                try:
                    if grid[y-1][x] == "|":
                        grid[y] = grid[y][:x]+"|"+grid[y][x+1:]
                        beams[y][x] += beams[y-1][x]
                except IndexError:
                    continue
    return [n_splits, grid, sum(beams[-1]), beams]
```

```python
_, _, beam_count, beams = propagate_beams(grid=input_data_pt2)
print(f"Total instances of beams reaching the end: {beam_count}")
# print(f"Overview of number of beams per position: {beams}")
```
