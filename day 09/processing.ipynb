```python
def get_area(positions: list[int]) -> list[float, int, int]:
    areas = []
    for one, c1 in enumerate(positions[:-1]):
        for two, c2 in enumerate(positions[one+1:]):
            areas.append(
                [
                    (abs(c1[0]-c2[0])+1)*(abs(c1[1]-c2[1])+1),
                    one,
                    one+two+1
                ]
            )
    return sorted(areas, key=lambda x: x[0], reverse=True)
```

```python
tile_positions = [[int(x) for x in line.split(",")] for line in input_data_pt1]
sorted_areas=get_area(positions=tile_positions)
print(f"Largest area found is {sorted_areas[0][0]} between tiles {sorted_areas[0][1]} and {sorted_areas[0][2]}")
```
